```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,50) 
```

```python
noise = 10 * np.random.randn(50,) # Creating noise with the random number generator
```

```python
y = 3 * x + 7 + noise
```

```python
plt.plot(x,y,'.');
```

```python
m_guess = 1 # Making the slope and intercept and plotting it out and making a guess for the slope
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
```

```python
m_guess = 3 # Real slope
b_guess = 7

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
```

```python
x= np.linspace(50)
m_guess = 50
b_guess = 100

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');

best_loss = 100000 
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,10,20):
    for b_guess in np.linspace(-10,10,20):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess)
```

```python
for m_guess in np.linspace(-10,10,10): # Trying a lot of different combonations for the Slope and Intercept and find the Loss 
    for b_guess in np.linspace(-10,10,10):
        
        y_guess = m_guess * x + b_guess

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,m_guess,b_guess)
```
